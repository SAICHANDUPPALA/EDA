```python
- step 1: find Q1:25  Q2:50  Q3:75
- step 2: IQR:  (Q3-Q1)
- step 3: Q1-1.5*IQR :Lower bound
- step 4: Q3+1.5*IQR :Upper bound
```

```python
Q1=np.quantile(data,0.25)
Q2=np.quantile(data,0.5)
Q3=np.quantile(data,0.75)
IQR=(Q3-Q1)
LB=Q1-1.5*IQR
UB=Q3+1.5*IQR
print('Q1:',Q1)
print('Q2:',Q2)
print('Q3:',Q3)
print('IQR:',IQR)
print('LB:',LB)
print('UB:',UB)
```

```python
#OTLIERS ARE LESS THAN LB
# outliers are gtreater than UB
#c1: data<LB
#c2: data>UB
data=visa_dataset['prevailing_wage']
con1=data<LB
con2=data>UB

con1|con2 #it will return bool values | is or
#in order to get data
#actual data: visadata
#we are extracting some part of data from main dataset by using conditions
#it will return a dataframe in columns
outlier_data=visa_dataset[con1|con2]
outlier_data
```
